### Import the required libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
```

###  Establish the needed Environmental & Viral Load Parameters

```python
room_sizes = [30, 60, 100]  # Room area in square meters
ventilation_rates = [1, 3, 6]  # Air changes per hour (ACH)
viral_loads = np.logspace(3, 9, num=10)  # From 1e3 to 1e9 particles/mL
duration_minutes = 60  # Exposure duration in minutes
```

### Transmission Simulation Function

```python
def estimate_infections(room_area, ventilation, viral_load, duration=60):
    exposure = viral_load / (ventilation * room_area) * duration
    infection_probability = exposure * 0.0001  # Adjust scaling here if needed
    infections = poisson.rvs(mu=infection_probability)
    return infections
```

### Running the Simulations Across Scenarios

```python
results = []

for room in room_sizes:
    for vent in ventilation_rates:
        for load in viral_loads:
            for i in range(10): 
                est_inf = estimate_infections(room, vent, load, duration_minutes)
                results.append({
                    'RoomSize_m2': room,
                    'Ventilation_ACH': vent,
                    'ViralLoad': load,
                    'EstimatedInfections': est_inf
                })

df = pd.DataFrame(results)

# Save to CSV if needed
df.to_csv("processed_room_scenarios12.csv", index=False)
print("✅ File 'processed_room_scenarios12.csv' has been saved successfully.")
```

### Visualizing the results of Infections vs Viral Load

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="ViralLoad", y="EstimatedInfections", hue="Ventilation_ACH", style="RoomSize_m2", palette="viridis")
plt.xscale("log")
plt.title("Estimated Infections vs Viral Load\n(Varying Ventilation and Room Size)")
plt.xlabel("Viral Load (log scale)")
plt.ylabel("Estimated Infections")
plt.legend(title="Ventilation (ACH)")
plt.tight_layout()
plt.show()
```

### Statistical Modeling using Poisson Regression

```python
df['LogViralLoad'] = np.log(df['ViralLoad'])
X = df[['LogViralLoad', 'RoomSize_m2', 'Ventilation_ACH']]
X = sm.add_constant(X)  # Adds intercept term
y = df['EstimatedInfections']

# Fit Poisson model
model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
print(model.summary())
```
